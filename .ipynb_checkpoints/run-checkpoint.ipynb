```python
import os
import math
import h5py
import numpy as np
import matplotlib as mpl
import grtrans_batch as gr
import matplotlib.pyplot as plt
from matplotlib.colors import TwoSlopeNorm
os.environ["OMP_NUM_THREADS"] = "1"
```

```python
kev = 2.41799050402417e17
ngrid = 100
a_spin = 0.99
mbh = 10
xlim = 20
```

```python
freq = 2*kev
ang = 50
amu = np.cos(ang*math.pi/180)
```

```python
x = np.linspace(-xlim, xlim, ngrid)
X, Y = np.meshgrid(x, x, indexing="ij")
```

```python
x = gr.grtrans()
x.write_grtrans_inputs('inputs.in',nfreq=1,nmu=1,mumin=amu,mumax=amu,fmin=freq,fmax=freq,fname="THINDISK",ename="BBPOL",nvals=4,spin=a_spin,mbh=mbh,nn=[ngrid,ngrid,1],gridvals=[-xlim, xlim,-xlim, xlim])
x.run_grtrans()
x.read_grtrans_output()
vals = x.ivals[:,:,0]
```

```python
stokesi = (vals[:,0].reshape((ngrid,ngrid)))
stokesq = (vals[:,1].reshape((ngrid,ngrid)))
stokesu = (vals[:,2].reshape((ngrid,ngrid)))
```

```python
# output #
outfile = "test.hdf5"
hf = h5py.File(outfile, 'w')

# create data #
hf.create_dataset('ngrid',data=ngrid)
hf.create_dataset('xlim',data=xlim)
hf.create_dataset('mbh',data=mbh)
hf.create_dataset('a_spin',data=a_spin)
hf.create_dataset('freq',data=freq)
hf.create_dataset('ang',data=ang)
hf.create_dataset('stokesi',data=stokesi.T)
hf.create_dataset('stokesq',data=stokesq.T)
hf.create_dataset('stokesu',data=stokesu.T)

# close #
hf.close()
```

```python
hf = h5py.File(outfile, 'r')
stokesi = hf['stokesi'][()]
```

```python
cmap = "afmhot"
vmin = stokesi.min()
vmax = stokesi.max()
norm = mpl.colors.Normalize(vmin=vmin, vmax=vmax)
tick = np.linspace(vmin, vmax, 10)
fig, ax = plt.subplots()
plt.contourf(X, Y, stokesi, 500, norm=norm, cmap=cmap)
plt.gca().set_aspect('equal')
plt.xlabel(r"$GM/c^{2}$")
plt.ylabel(r"$GM/c^{2}$")
cbar_ax = fig.add_axes([0.84, 0.15, 0.03, 0.7])
cbar = mpl.colorbar.ColorbarBase(cbar_ax, cmap=cmap, norm=norm, ticks=tick)
plt.show()
```

```python
cmap = "bwr"
vmin = -stokesq.max()
vcen = 0
vmax = stokesq.max()
norm = TwoSlopeNorm(vmin=vmin, vcenter=vcen, vmax=vmax)
tick1 = np.linspace(vmin, vcen, 5, endpoint=False)
tick2 = np.linspace(vcen, vmax, 5, endpoint=True)
tick = np.concatenate((tick1, tick2), axis=0)
fig, ax = plt.subplots()
plt.contourf(X, Y, stokesq, 2000, norm=norm, cmap=cmap)
plt.gca().set_aspect('equal')
plt.xlabel(r"$GM/c^{2}$")
plt.ylabel(r"$GM/c^{2}$")
cbar_ax = fig.add_axes([0.84, 0.15, 0.03, 0.7])
cbar = mpl.colorbar.ColorbarBase(cbar_ax, cmap=cmap, norm=norm, ticks=tick)
plt.show()
```

```python
cmap = "coolwarm"
vmin = stokesu.min()
vcen = 0
vmax = -stokesu.min()
norm = TwoSlopeNorm(vmin=vmin, vcenter=vcen, vmax=vmax)
tick1 = np.linspace(vmin, vcen, 5, endpoint=False)
tick2 = np.linspace(vcen, vmax, 5, endpoint=True)
tick = np.concatenate((tick1, tick2), axis=0)
fig, ax = plt.subplots()
plt.contourf(X, Y, stokesu, 2000, norm=norm, cmap=cmap)
plt.gca().set_aspect('equal')
plt.xlabel(r"$GM/c^{2}$")
plt.ylabel(r"$GM/c^{2}$")
cbar_ax = fig.add_axes([0.84, 0.15, 0.03, 0.7])
cbar = mpl.colorbar.ColorbarBase(cbar_ax, cmap=cmap, norm=norm, ticks=tick)
plt.show()
```

```python
LP = np.sqrt(stokesq**2 + stokesu**2)/stokesi*100
```

```python
cmap = "magma"
vmin = np.nanmin(LP)
vmax = np.nanmax(LP)
norm = mpl.colors.Normalize(vmin=vmin, vmax=vmax)
tick = np.linspace(vmin, vmax, 10)
fig, ax = plt.subplots()
plt.contourf(X, Y, LP, 500, norm=norm, cmap=cmap)
plt.gca().set_aspect('equal')
plt.xlabel(r"$GM/c^{2}$")
plt.ylabel(r"$GM/c^{2}$")
cbar_ax = fig.add_axes([0.84, 0.15, 0.03, 0.7])
cbar = mpl.colorbar.ColorbarBase(cbar_ax, cmap=cmap, norm=norm, ticks=tick)
plt.show()
```

```python
evpa = (180/math.pi)*0.5*np.arctan2(stokesu,stokesq)
```

```python
cmap = "magma"
vmin = np.nanmin(LP)
vmax = np.nanmax(LP)
norm = mpl.colors.Normalize(vmin=vmin, vmax=vmax)
tick = np.linspace(vmin, vmax, 10)
fig, ax = plt.subplots()
plt.contourf(X, Y, LP, 500, norm=norm, cmap=cmap)
vxp = LP/vmax*np.sin(evpa*math.pi/180)
vyp = - LP/vmax*np.cos(evpa*math.pi/180)
skip = int(ngrid/32) 
plt.quiver(X[::skip,::skip], Y[::skip,::skip], vxp[::skip,::skip], vyp[::skip,::skip], 
           headwidth=1, headlength=1, width=0.002, color='#00ff00', units='width', scale=20, pivot='mid')
plt.gca().set_aspect('equal')
plt.xlabel(r"$GM/c^{2}$")
plt.ylabel(r"$GM/c^{2}$")
cbar_ax = fig.add_axes([0.84, 0.15, 0.03, 0.7])
cbar = mpl.colorbar.ColorbarBase(cbar_ax, cmap=cmap, norm=norm, ticks=tick)
plt.show()
```
